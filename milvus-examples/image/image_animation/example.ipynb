```python
import requests
import utils
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

file_name = './test.png' # Image File path to be sent

url_path = "http://127.0.0.1:8000" # Base IP address of API
input_image = Image.open(file_name)
response = requests.post(
    f"{url_path}/api/trans",files={"file":open(file_name,"rb")}, data={"model_name": "celeba"}
    )


output_image = utils.read_image(response.content)
```

```python
input_arr = np.array(input_image)
output_arr = np.array(output_image)
plt.figure(figsize = (12,12))
plt.subplot(121)
plt.imshow(input_arr)
plt.title("Input Image")
plt.axis('off')
plt.subplot(122)
plt.imshow(output_arr)
plt.title("Transformed Image")
plt.axis('off')
```
